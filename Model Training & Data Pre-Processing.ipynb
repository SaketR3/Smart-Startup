# **Data Splitting**   

```python
path = os.path.join(DATADIR, 'cleaned_startup_loan_data.csv')
df = pd.read_csv(path)
df = df.drop(columns=['Unnamed: 0'])
df
```

```python
X_train_init = df.drop(columns=['MIS_Status'])
X_train_init
```

```python
binencoder = LabelBinarizer()
y_train_init = df['MIS_Status']
y_train_init = y_train_init.astype(int)
y_train_init = binencoder.fit_transform(y_train_init)
y_train_init
```

```python
X_train_full, X_test, y_train_full, y_test = train_test_split(X_train_init,
                                                    y_train_init,
                                                    test_size=0.1,
                                                    shuffle=True,
                                                    random_state=42)
```

```python
X_train_full.shape
```

```python
X_train_df, y_train = X_train_full[:-20000], y_train_full[:-20000]
X_valid_df, y_valid = X_train_full[-20000:], y_train_full[-20000:]
```

```python
X_train_df.shape
```

# **Model Training**

```python
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(),

    tf.keras.layers.Dense(100, activation='swish',
                              kernel_initializer='he_normal'),

    tf.keras.layers.Dense(100, activation='swish',
                              kernel_initializer='he_normal'),

    tf.keras.layers.Dense(100, activation='swish',
                              kernel_initializer='he_normal'),

    tf.keras.layers.Dense(100, activation='swish',
                              kernel_initializer='he_normal'),

    tf.keras.layers.Dense(100, activation='swish',
                              kernel_initializer='he_normal'),

    tf.keras.layers.Dense(1, activation='sigmoid'),
])
```

```python
model.compile(
    loss="binary_crossentropy",
    optimizer=tf.keras.optimizers.Nadam(),
    metrics=["accuracy"],
)
```

```python
print(model.optimizer.learning_rate.numpy())
```

```python
model.optimizer.learning_rate.assign(1e-05)
```

```python
print(model.optimizer.learning_rate.numpy())
```

Note: the model was trained for around 25 epochs; only the last 5 epochs are displayed in the output. 

```python
history = model.fit(
    X_train_df.to_numpy(),
    y_train,
    validation_data=(X_valid_df, y_valid),
    epochs=5,
)
```

```python
model.evaluate(X_test, y_test)
```

```python
X_new = X_test[:20]
y_proba = model.predict(X_new)
print(y_proba)
print(y_test[:20])
```
